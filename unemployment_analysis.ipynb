```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
#CSV file load
data = pd.read_csv("Unemployment in India.csv")
print("Data Loaded Successfully!\n")
#first few Rows
print("First 5 Rows:\n")
print(data.head())
#column names check 
print("\nColumns in Dataset:")
print(data.columns)

#make short and simple column name 
data.rename(columns={
    'Region': 'State',
    ' Date': 'Date',
    ' Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    ' Estimated Employed': 'Employed',
    ' Estimated Labour Participation Rate (%)': 'Labour_Participation'
}, inplace=True)
#check missing values
print("\nMissing Values:\n")
print(data.isnull().sum())
# convert date to datetime format
data['Date'] = pd.to_datetime(data['Date'],
                             dayfirst=True)
#plotting unemployment rate by state
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Unemployment_Rate', hue='State', data=data)
plt.title("Unemployment Rate Over Time by State")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
#avg unemployment by state
avg_unemp = data.groupby('State')['Unemployment_Rate'].mean()

print("\nAverage Unemployment Rate of each State:\n")
print(avg_unemp)
plt.figure(figsize=(10, 6))
avg_unemp.plot(kind='bar',color='skyblue')
plt.title("Average Unemployment Rate by State")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
